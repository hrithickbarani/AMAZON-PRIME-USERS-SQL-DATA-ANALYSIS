# USER DEMOGRAPHICS

## 1. What is the age distribution of users?

```python
pd.read_sql_query('''SELECT 
  CASE
    WHEN TIMESTAMPDIFF(YEAR, date_of_birth, CURDATE()) BETWEEN 18 AND 25 THEN '18-25'
    WHEN TIMESTAMPDIFF(YEAR, date_of_birth, CURDATE()) BETWEEN 26 AND 35 THEN '26-35'
    WHEN TIMESTAMPDIFF(YEAR, date_of_birth, CURDATE()) BETWEEN 36 AND 45 THEN '36-45'
    WHEN TIMESTAMPDIFF(YEAR, date_of_birth, CURDATE()) BETWEEN 46 AND 55 THEN '46-55'
    WHEN TIMESTAMPDIFF(YEAR, date_of_birth, CURDATE()) > 55 THEN '56+'
    ELSE 'Unknown'
  END AS age_group,
  COUNT(*) AS user_count
FROM amazon_prime_users
GROUP BY age_group
ORDER BY age_group;
''', engine)
```

## *Insights:*

- **18-25 years:** 223 users (8.92%)  
- **26-35 years:** 375 users (15.00%)  
- **36-45 years:** 348 users (13.92%)  
- **46-55 years:** 341 users (13.64%)  
- **56+ years:** 1213 users (48.52%)

The majority of users belong to the **56+ years** age group, indicating a stronger user base among older individuals. Younger users (18-25) constitute the smallest proportion. This suggests potential to attract a younger demographic to balance the user base.

# SUBSCRIPTION ANALYSIS

## 4. How many users are subscribed to each subscription plan?

```python
pd.read_sql_query('''
    SELECT subscription_plan, COUNT(user_id) AS user_count
    FROM amazon_prime_users
    GROUP BY subscription_plan;
''', engine)
```

## *Insights:*
- **Annual Subscription**: 1,271 users are subscribed to the annual plan, indicating a significant portion of users prefer long-term subscriptions.
- **Monthly Subscription**: 1,229 users are subscribed to the monthly plan, which suggests a slightly lower, but still significant preference for flexibility in subscription duration.


## 5. What is the churn rate (users who didnt renew their membership)?  

```python
query='''WITH UserStatus AS (
    SELECT 
        COUNT(*) AS total_users,
        SUM(CASE WHEN membership_end_date < CURDATE() THEN 1 ELSE 0 END) AS churned_users
    FROM amazon_prime_users
)
SELECT 
    churned_users / total_users AS churn_rate
FROM UserStatus;
'''
pd.read_sql_query(query, engine)
```

## *Insights:*

- **Churn Rate:** **11.36%**
- **Calculation:** (Churned Users / Total Users)

**11.36% of users have churned, highlighting the need for improved retention strategies.**


## 6. What is the average duration of user subscriptions?

```python
pd.read_sql_query('''
SELECT 
    AVG(DATEDIFF(membership_end_date, membership_start_date)) AS avg_subscription_duration
FROM 
    amazon_prime_users;
''', engine)
```

## *Insights:*

- The average subscription duration for users is 365 days, which indicates that most users are subscribed for a full year. This suggests that users tend to choose the annual subscription plan, which may be more popular due to its cost-effectiveness.


# USER ENGAGEMENT

## 7. What is the average engagement score for each subscription plan?  

```python
pd.read_sql_query('''
SELECT 
    subscription_plan,
    AVG(CASE 
        WHEN engagement_metrics = 'low' THEN 1
        WHEN engagement_metrics = 'medium' THEN 2
        WHEN engagement_metrics = 'high' THEN 3
        ELSE NULL 
    END) AS avg_engagement
FROM amazon_prime_users
GROUP BY subscription_plan;
''', engine)
```

## *Insight:*
- **Average Engagement Score**:
  - Users on the **Monthly** subscription plan have a slightly higher average engagement score (2.01) compared to those on the **Annual** plan (2.01).
  - Both plans have engagement scores that suggest the majority of users fall into the **"medium"** category, indicating moderate engagement with the platform.

This could suggest that both groups engage with the platform similarly, with only a minor difference between the plans.

# TREND ANALYSIS

## 15. How has the number of new users (based on membership dates) changed over time?

```python
pd.read_sql_query('''
SELECT 
    EXTRACT(YEAR_MONTH FROM membership_start_date) AS month,
    COUNT(*) AS new_users
FROM amazon_prime_users
GROUP BY month
ORDER BY month;
''', engine)
```

## *Insights:*
- The number of new users joining the platform shows a slight fluctuation across months. 
- **January 2024** saw the highest number of new users at **773**, while **April 2024** had the lowest count with **332** new users.


## 16. Are certain months or seasons associated with higher subscription rates?

```python
pd.read_sql_query('''
SELECT 
    CASE 
        WHEN MONTH(membership_start_date) IN (12, 1, 2) THEN 'Winter'
        WHEN MONTH(membership_start_date) IN (3, 4, 5) THEN 'Spring'
        WHEN MONTH(membership_start_date) IN (6, 7, 8) THEN 'Summer'
        WHEN MONTH(membership_start_date) IN (9, 10, 11) THEN 'Fall'
    END AS season,
    COUNT(*) AS new_subscriptions
FROM amazon_prime_users
GROUP BY season
ORDER BY FIELD(season, 'Winter', 'Spring', 'Summer', 'Fall');
''', engine)
```

## *Insights:*


- **Winter** has the highest number of new subscriptions, with **1,424** users signing up.
- **Spring** follows with **1,076** new subscriptions.
